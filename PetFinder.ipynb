```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
import glob
import cv2
import os
```

```python
def load_pawpularity_images(inputPath):
	images = []
	count=0
	for filename in glob.glob(os.path.join(inputPath, "*.jpg")):
		img = cv2.imread(filename)
		image = cv2.resize(img, (64, 64))
		images.append(image)
		count = count +1
	print(count)
	return np.array(images)
```
